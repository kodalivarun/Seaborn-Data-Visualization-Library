### Box Plot

##### A box plot (or box-and-whisker plot)  shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
from IPython.display import Image
Image(filename='1_2c21SkzJMf3frPXPAR_gZA.png')
```

### The summary we can extract from a boxplot are:
Quartiles, specifically the first and third quartiles, which correspond to the 25th and 75th percentiles.

Median, the mid-point in the distribution, which also corresponds to the 50th percentile.

Interquartile range (IQR), the width between the third and first quantiles. Expressed 
mathematically, we have IQR = Q3 — Q1.

lower bound, minimum value in the dataset excluding outliers, which corresponds to Q1–1.5xIQR.

upper bound, maximum value in the dataset, excluding outliers, which corresponds to Q3+ 1.5xIQR.

(data point value) < Q1–1.5xIQR, then it’s an outlier.
(data point value) > Q3 + 1.5xIQR, then it’s an outlier.

# Heat Map

### In order for a heatmap to work properly, your data should already be in a matrix form, and what i mean by matrix form is that the index name and the column name should match up so that the cell value actually indicates something that is relevant to both of those names......the sns.heatmap function basically colors it in for you. 

### we can covert the data into matrix form in two ways:  by correlation method and by using pivot table

```python
# Matrix form for correlation data
df1.corr()
```

### now we have matrix form...if you see the matrix table here now we have column and index  actually indicates a relevant value where they match up and this is the matrix form you gonna need your data to be in to apply heatmap on this

### one more key point here is that correlation & pivot table can be applied only on the numerical columns 

```python
sns.heatmap(df1.corr())
```

### heatmap colors all those values based on some sort of gradient scale.

So if I take a look at my original data frame it essentially just set these numbers to a gradient on a color scale and then fill that in.

And that way if you have a very large correlation data frame or some other matrix form data you can easily tell what the numbers are what the values are relative to each other using a heat map.

correlation basically helps us to find out interelation b/w all the columns in our dataset

And that's the main primary use for a matrix plot.

### working with markers

### lmplot kwargs get passed through to regplot which is a more general form of lmplot(). regplot has a scatter_kws parameter that gets passed to plt.scatter. So you want to set the s parameter in that dictionary, which corresponds (a bit confusingly) to the squared markersize. In other words you end up passing a dictionary with the base matplotlib arguments, in this case, s for size of a scatter plot. In general, you probably won't remember this off the top of your head, but instead reference the documentation.

```python
# http://matplotlib.org/api/markers_api.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex'
           ,palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```
